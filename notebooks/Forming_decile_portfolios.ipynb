```python

crsp["decile"] = (
    crsp.groupby("date", group_keys=False)
    .apply(lambda df: pd.qcut(df["momentum"], 10, labels=False, duplicates="drop"))
)



decile_returns = (
    crsp.groupby(["date", "decile"])["ret"]
    .mean()
    .unstack()
)



long_short = decile_returns[9] - decile_returns[0]


print("Momentum preview:")
print(crsp[["permno", "date", "momentum"]].head())

print("\nDecile returns preview:")
print(decile_returns.head())

print("\nLong-short returns preview:")
print(long_short.head())
```
